## Question 3

_How many taxi trips were there on January 15?_

```python
len(df[(df['tpep_pickup_datetime'].dt.month == 1) & (df['tpep_pickup_datetime'].dt.day == 15)])
```

## Question 4

_Find the largest tip for each day. On which day it was the largest tip in January?_

```python
df.iloc[df['tip_amount'].argmax()]['tpep_pickup_datetime'].date()
```

## Question 5

_What was the most popular destination for passengers picked up in central park on January 14?_

```python
df_day14 = df[(df['tpep_pickup_datetime'].dt.month == 1) & (df['tpep_pickup_datetime'].dt.day == 14)]
loc_id_central_park = df_zone[df_zone['Zone'] == 'Central Park'].iloc[0]['LocationID']
loc_id_most_popular = df_day14[df_day14['PULocationID'] == loc_id_central_park]['DOLocationID'].value_counts().index[0]
df_zone[df_zone['LocationID'] == loc_id_most_popular].iloc[0]['Zone']
```

## Question 6

_What's the pickup-dropoff pair with the largest average price for a ride (calculated based on `total_amount`)?_

```python
df.groupby(['PULocationID', 'DOLocationID'])['total_amount'].mean().sort_values(ascending=False)
```
